```python
nc = Dataset('/Users/kuangdai/Downloads/axisem3d_synthetics_test_100.nc')
```

```python
# com='ENZ'
com='RTZ.strain'
data0 = nc.variables['II.ST1.' + com][:,:]
data1 = nc.variables['II.ST2.' + com][:,:]
data2 = nc.variables['II.ST3.' + com][:,:]
data3 = nc.variables['II.ST4.' + com][:,:]
```

```python
chi0 = data0[:, 0]
chi1 = data1[:, 0]
chi2 = data2[:, 0]
chi3 = data3[:, 0]
```

```python
# use this file for stress output
# ~/Cartesian_Developer/AxiSEM3D/SOLVER/src/core/element/FluidElement.cpp
```

```python
time = nc.variables['time_points'][:]
stats = Stats()
stats.starttime = UTCDateTime(time[0])
stats.delta = time[1] - time[0]
stats.npts = len(time)
```

```python
trace0 = Trace(chi0, header=stats)
trace1 = Trace(chi1, header=stats)
trace2 = Trace(chi2, header=stats)
trace3 = Trace(chi3, header=stats)
```
